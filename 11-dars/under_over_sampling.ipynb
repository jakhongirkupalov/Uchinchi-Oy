```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from imblearn.under_sampling import NearMiss, RandomUnderSampler, ClusterCentroids

x, y = make_classification(n_samples=500, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9, 0.1], random_state=42)

# original class destribution
plt.figure(figsize=(18,10))
plt.subplot(2,3,1)
plt.scatter(x[y==0][:,0], x[y==0][:,1], label="Majority", alpha=0.5)
plt.scatter(x[y==1][:,0], x[y==1][:,1], label="Minority", alpha=0.5, color="red")
plt.title("original dataset")
plt.legend()


# RandomOverSampler

ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x,y)
plt.subplot(2,3,2)
plt.scatter(x_ros[y_ros==0][:,0], x_ros[y_ros==0][:,1], label="Majority", alpha=0.5)
plt.scatter(x_ros[y_ros==1][:,0], x_ros[y_ros==1][:,1], label="Minority", alpha=0.5, color="red")
plt.title("Random Over Sampler")
plt.legend()

# SMOTE

smote = SMOTE(random_state=42)
x_smote, y_smote = smote.fit_resample(x,y)
plt.subplot(2,3,3)
plt.scatter(x_smote[y_smote==0][:,0], x_smote[y_smote==0][:,1], label="Majority", alpha=0.5)
plt.scatter(x_smote[y_smote==1][:,0], x_smote[y_smote==1][:,1], label="Minority", alpha=0.5, color="red")
plt.title("SMOTE")
plt.legend()

#RandomUNderSampling

rus = RandomUnderSampler(random_state=42)
x_rus, y_rus = rus.fit_resample(x,y)
plt.subplot(2,3,4)
plt.scatter(x_rus[y_rus==0][:,0], x_rus[y_rus==0][:,1], label="Majority", alpha=0.5)
plt.scatter(x_rus[y_rus==1][:,0], x_rus[y_rus==1][:,1], label="Minority", alpha=0.5, color="red")
plt.title("Rundom UNder Sampling")
plt.legend()


#nearmis
nearmis = NearMiss(version=1)
x_nearmis, y_nearmis = nearmis.fit_resample(x,y)
plt.subplot(2,3,5)
plt.scatter(x_nearmis[y_nearmis==0][:,0], x_nearmis[y_nearmis==0][:,1], label="Majority", alpha=0.5)
plt.scatter(x_nearmis[y_nearmis==1][:,0], x_nearmis[y_nearmis==1][:,1], label="Minority", alpha=0.5, color="red")
plt.title("NEAR MISS")
plt.legend()


# Cluster Centroids
cc = ClusterCentroids(random_state=42)
x_cc, y_cc = cc.fit_resample(x,y)
plt.subplot(2,3,6)
plt.scatter(x_cc[y_cc==0][:,0], x_cc[y_cc==0][:,1], label="Majority", alpha=0.5)
plt.scatter(x_cc[y_cc==1][:,0], x_cc[y_cc==1][:,1], label="Minority", alpha=0.5, color="red")
plt.title("ClusterCentroids")
plt.legend()


plt.tight_layout()
plt.show()
```

# OVERSAMPLING

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from collections import Counter

df = pd.read_csv(r"C:\Users\Jahongir\desktop\practise\Data\Row_Data\creditcard.csv")

x = df.drop("Class", axis=1)
y= df["Class"]

print("Original class distribution: ", Counter(y))

#RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x,y)
print("RandomOverSampler", Counter(y_ros))

#SMOTE
smote = SMOTE(random_state=42)
x_smote, y_smote = smote.fit_resample(x,y)
print("SMOTE", Counter(y_smote))

#ADASYN
adasyn = ADASYN(random_state=42)
x_adasyn, y_adasyn = adasyn.fit_resample(x,y)
print("ADASYN", Counter(y_adasyn))
```
