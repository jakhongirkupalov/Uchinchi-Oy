```python
import pandas as pd

# 'Published Date' ustunini datetime formatiga o'tkazish
df["Published Date"] = pd.to_datetime(df["Published Date"],errors="coerce",utc=True).dt.tz_convert(None)

# Yil, oy, hafta va kun kabi yangi ustunlarni yaratish
df["pub_year"] = df["Published Date"].dt.year
df["pub_month"] = df["Published Date"].dt.month
df["pub_week"] = df["Published Date"].dt.isocalendar().week
df["pub_day"] = df["Published Date"].dt.day
df["pub_weekday"] = df["Published Date"].dt.weekday

# Bugungi sana bilan nashr qilingan sana orasidagi farqni kunlarda hisoblash
df["days_since_pub"] = (pd.Timestamp.today(tz=None) - df["Published Date"]).dt.days
```

```python
df["like_ratio"] = df["Likes"] / (df["Views"] + 1)
df["coment_ratio"] = df["Comments"] / (df["Views"] + 1)

df["views_per_sub"] = df["Views"] / (df["Subscribers"] + 1)
df["likes_per_sub"] = df["Likes"] / (df["Subscribers"] + 1)
```
